```python
# A Game of Tic-Tac-Toe

# Displaying the board first
ttt_display_list = ['_|_|_','_|_|_',' | | ']

def TTT_GAME_BOARD():
    for i in ttt_display_list:
        print(i)

print("The cure to boredom is TIC-TAC-TOE!")
TTT_GAME_BOARD()

# Choice of symbol and player
player_1 = 'X'
player_2 = 'O'

player_choice = input("Choose your player:\nPlayer_1 -> 'X'\nPlayer_2 -> 'O'\nNote: Please enter your choice in the format: X or O")


while (player_choice != 'X') and (player_choice != 'O'):
    print('\nChoose an appropriate symbol first in order to start the game')
    player_choice = input("\nChoose your player:\nPlayer_1 -> 'X'\nPlayer_2 -> 'O'\nNote: Please enter your choice in the format: X or O")
    if (player_choice == 'X') or (player_choice == 'O'):
        break
    else:
        continue
    
# The game proceeds only if a suitable symbol is choosen. Else, it stops and asks for a suitable input
if (player_choice == 'X') or (player_choice == 'O'):
    # List of positions available to the first mover
    position_list = []

    for i in range(3):
        for j in range(3):
            position_list.append([i,j])

    # The game progression
    rem_list = [] # rem_list implies the list containing the split of the element in the display list to be removed
    # By default counter is set to zero. It becomes 1 when either of the players win
    counter = 0

    for m in range(1,10):
        if ((m%2 != 0) and (player_choice == 'X')) or ((m%2 == 0) and (player_choice == 'X')):
            inp_ = input("\nPlayer_1 (X) please make a move\nPlease select a position out of this list in the format i,j\n{}".format(position_list))
            inp_1 = int(inp_.split(',')[0])
            inp_2 = int(inp_.split(',')[1])
            move_list = list([inp_1,inp_2])
            # A player must enter a valid position. In case an invalid position is entered, it keeps on asking for a valid position until one is entered
            while move_list not in position_list:
                print('\nPosition entered does not exist or has been taken up')
                inp_ = input("\nPlease select a position from the list below\n{}".format(position_list))
                inp_1 = int(inp_.split(',')[0])
                inp_2 = int(inp_.split(',')[1])
                move_list = list([inp_1,inp_2])                
                if move_list in position_list:
                    break
                else:
                    continue
            position_list.remove(move_list)
            replaced_list = []
            for i in range(3): 
                r_list_elem = ttt_display_list[move_list[0]].split('|')[i]
                replaced_list.append(r_list_elem)

            replaced_list.pop(move_list[1])
            replaced_list.insert(move_list[1],'X')
            replaced_list

            replaced_string = replaced_list[0]+'|'+replaced_list[1]+'|'+replaced_list[2]

            # Updating the board display depending on the player's move
            if move_list[0] == 0:
                ttt_display_list.pop(0)
                ttt_display_list.insert(0,replaced_string)
            elif move_list[0] == 1:
                ttt_display_list.pop(1)
                ttt_display_list.insert(1,replaced_string)
            else:
                ttt_display_list.pop(2)    
                ttt_display_list.insert(2,replaced_string)


            # Conditions checking for a win or a draw
            if (ttt_display_list[0] == 'X|X|X') or (ttt_display_list[1] == 'X|X|X') or (ttt_display_list[2] == 'X|X|X'):
                print("\nPlayer_1 has won!")
                counter = 1
                TTT_GAME_BOARD()
                break
            elif ((ttt_display_list[0][0] == 'X') and (ttt_display_list[1][2] == 'X') and (ttt_display_list[2][4] == 'X')):
                print("\nPlayer_1 has won!")
                counter = 1
                TTT_GAME_BOARD()
                break
            elif ((ttt_display_list[0][4] == 'X') and (ttt_display_list[1][2] == 'X') and (ttt_display_list[2][0] == 'X')):
                print("\nPlayer_1 has won!")
                counter = 1
                TTT_GAME_BOARD()
                break
            elif ((ttt_display_list[0][0] == 'X') and (ttt_display_list[1][0] == 'X') and (ttt_display_list[2][0] == 'X')) or ((ttt_display_list[0][1] == 'X') and (ttt_display_list[1][1] == 'X') and (ttt_display_list[2][1] == 'X')) or ((ttt_display_list[0][2] == 'X') and (ttt_display_list[1][2] == 'X') and (ttt_display_list[2][2] == 'X')):
                print("\nPlayer_1 has won!")
                counter = 1
                TTT_GAME_BOARD()
                break
            else:
                TTT_GAME_BOARD()

            if (m == 9) and (counter == 0):
                print("\nSadly, it's a tie. Play again to win!")
            else:
                pass
            player_choice = 'O'
        # The same game progression is followed for the second player
        else:
            inp_ = input("\nPlayer_2 (0) please make a move\nPlease select a position out of this list in the format i,j\n{}".format(position_list))
            inp_1 = int(inp_.split(',')[0])
            inp_2 = int(inp_.split(',')[1])
            move_list = list([inp_1,inp_2])
            while move_list not in position_list:
                print('\nPosition entered does not exist or has been taken up')
                inp_ = input("\nPlease select a position from the list below\n{}".format(position_list))
                inp_1 = int(inp_.split(',')[0])
                inp_2 = int(inp_.split(',')[1])
                move_list = list([inp_1,inp_2])                
                if move_list in position_list:
                    break
                else:
                    continue
            position_list.remove(move_list)
            replaced_list = []
            for i in range(3): 
                r_list_elem = ttt_display_list[move_list[0]].split('|')[i]
                replaced_list.append(r_list_elem)

            replaced_list.pop(move_list[1])
            replaced_list.insert(move_list[1],'O')
            replaced_list

            replaced_string = replaced_list[0]+'|'+replaced_list[1]+'|'+replaced_list[2]

            if move_list[0] == 0:
                ttt_display_list.pop(0)
                ttt_display_list.insert(0,replaced_string)
            elif move_list[0] == 1:
                ttt_display_list.pop(1)
                ttt_display_list.insert(1,replaced_string)
            else:
                ttt_display_list.pop(2)    
                ttt_display_list.insert(2,replaced_string)



            if (ttt_display_list[0] == 'O|O|O') or (ttt_display_list[1] == 'O|O|O') or (ttt_display_list[2] == 'O|O|O'):
                print("\nPlayer_2 has won!")
                TTT_GAME_BOARD()
                counter = 1
                break
            elif ((ttt_display_list[0][0] == 'O') and (ttt_display_list[1][2] == 'O') and (ttt_display_list[2][4] == 'O')):        
                print("\nPlayer_2 has won!")
                counter = 1
                TTT_GAME_BOARD()
                break
            elif ((ttt_display_list[0][4] == 'O') and (ttt_display_list[1][2] == 'O') and (ttt_display_list[2][0] == 'O')):
                print("\nPlayer_2 has won!")
                counter = 1
                TTT_GAME_BOARD()
                break
            elif ((ttt_display_list[0][0] == 'O') and (ttt_display_list[1][0] == 'O') and (ttt_display_list[2][0] == 'O')) or ((ttt_display_list[0][1] == 'O') and (ttt_display_list[1][1] == 'O') and (ttt_display_list[2][1] == 'O')) or ((ttt_display_list[0][2] == 'O') and (ttt_display_list[1][2] == 'O') and (ttt_display_list[2][2] == 'O')):
                print("\nPlayer_2 has won!")
                counter = 1
                TTT_GAME_BOARD()
                break
            else:
                TTT_GAME_BOARD()

            if (m == 9) and (counter == 0):
                print("\nSadly, it's a tie. Play again to win!")
                
            player_choice = 'X'

else:
    print("\nPlease give a suitable input first")  
```
